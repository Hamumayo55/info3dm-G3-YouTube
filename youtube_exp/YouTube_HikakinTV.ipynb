```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#散布図行列と相関行列の作成

cols = ['viewCount','likeCount','dislikeCount','len_title']

sns.pairplot(df[cols], size=2.5)
plt.tight_layout()
# plt.savefig('images/10_03.png', dpi=300)
plt.show()

cm = np.corrcoef(df[cols].values.T)
#sns.set(font_scale=1.5)
hm = sns.heatmap(cm,
                 cbar=True,
                 annot=True,
                 square=True,
                 fmt='.2f',
                 annot_kws={'size': 15},
                 yticklabels=cols,
                 xticklabels=cols)

plt.tight_layout()
# plt.savefig('images/10_04.png', dpi=300)
plt.show()
```

```python
#データの準備
df_explain = df[['likeCount','dislikeCount','len_title']]
df_target = df[['viewCount']]
```

```python
#線形回帰
from sklearn import linear_model
clf = linear_model.LinearRegression() 
```

```python
clf.fit(df_explain,df_target)
```

```python
print('intercept b = ', clf.intercept_)
```

```python
from sklearn.metrics import mean_squared_error
y_pred = clf.predict(df_explain)
mse = mean_squared_error(df_target, y_pred)
print('MSE = ', mse)
```

```python
print('R^2 = ', clf.score(df_explain,df_target))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_explain,df_target, random_state=0)
```

```python
print('train = ', len(X_train))
print('test =', len(X_test))
```

```python
from sklearn import linear_model
clf = linear_model.LinearRegression()

clf.fit(X_train, y_train)
```

```python
print(clf.score(X_train, y_train))
print(clf.score(X_test, y_test))
```

```python
from sklearn.model_selection import cross_val_score

clf = linear_model.LinearRegression() 
cvl = cross_val_score(clf,X_train, y_train, cv=5)
print('Cross-Validation scores: {}'.format(cvl))
# スコアの平均値
print('Average score: {}'.format(np.mean(cvl)))
```

```python
df_OneMillion = df[df["viewCount"] > 3000000]
```

```python
#形態素解析
from janome.tokenizer import Tokenizer
import collections

list_tword = []
t = Tokenizer()
for i in range(len(df_OneMillion)):
    title = df_OneMillion.iloc[i,1]
    for token in t.tokenize(title):
        list_tword.append((token.surface))
        c = collections.Counter(list_tword)
print(c)
```

```python
#[]が含まれるなら1,含まれないなら0

buzz_word_Braces = ["【","】",]
for i in buzz_word_Braces:
    df.loc[df['title'].str.contains(i),"title_encode_Braces"] = 1
df.loc[df['title_encode_Braces'].isnull(),"title_encode_Braces"] = 0
```

```python
#バズりやすい言葉があれば1,それ以外を0

buzz_word = ["超","円","'www","巨大","開封"]
for n in buzz_word:
    df.loc[df['title'].str.contains(n),"title_encode_word"] = 1
df.loc[df['title_encode_word'].isnull(),"title_encode_word"] = 0
```

```python
#散布図行列と相関行列の作成

cols = ['viewCount','likeCount','dislikeCount','len_title','title_encode_Braces','title_encode_word']

sns.pairplot(df[cols], size=2.5)
plt.tight_layout()
# plt.savefig('images/10_03.png', dpi=300)
plt.show()

cm = np.corrcoef(df[cols].values.T)
#sns.set(font_scale=1.5)
hm = sns.heatmap(cm,
                 cbar=True,
                 annot=True,
                 square=True,
                 fmt='.2f',
                 annot_kws={'size': 10},
                 yticklabels=cols,
                 xticklabels=cols)

plt.tight_layout()
# plt.savefig('images/10_04.png', dpi=300)
plt.show()
```
